```python
import pandas as pd
import matplotlib.pyplot as plt
import time
start = time.time()
df= pd.read_csv('Datafiles/marksnw.csv')
x=df.values
x=[]
failed=0
d=0
f=0
s=0
t=0
for row in df.values:
    x.append((row[1]))
for i in range(len(x)):    
        if(x[i]>80):
            d=d+1
        elif(x[i]>70) and (x[i]<=80):
            f=f+1
        elif(x[i]>60) and (x[i]<=70):
            s=s+1
        elif(x[i]>50) and (x[i]<=60):
            t=t+1
        else:
            failed=failed+1
labels = 'distinction', 'first', 'second', 'third','fail'
sizes = [d, f, s, t,failed]
explode = (0, 0, 0, 0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
end = time.time()
print(end-start)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
start = time.time()
df= pd.read_csv('datafiles/AppleStore.csv')
y=[]
cnt=[]
labels=[]
sizes=[]
for row in df.values:
    y.append(row[12])
un=np.unique(y)
for i in range(0,len(un)):
    count=0
    for j in range(len(y)):
        if(un[i]==y[j]):
            count=count+1
    cnt.append(count)
#print(cnt)
for l in range(len(un)):
    labels.append(un[l])
for s in range(len(cnt)):
    sizes.append(cnt[s])
#sizes = [d, f, s, t,failed]
explode = (1, 1, 1, 1,1) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()       
end = time.time()
print(end-start)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt;plt.rcdefaults()
import time
start = time.time()
df= pd.read_csv('Datafiles/AppleStore.csv')
df=df[['prime_genre','price','user_rating']]
df=df[df.prime_genre=='Sports']
free=[]
paid=[]
for row in df.values:
    if(row[1]==0.00):
        free.append(row[2])
    else:
        paid.append(row[2])
#print(free)
#print(paid)
x=sum(free)/len(free)
y=sum(paid)/len(paid)

objects=('Free average', 'Paid average')
y_pos=np.arange(len(objects))
performance=[x,y]
plt.bar(y_pos,performance,width=0.5, color="pink")
plt.xticks(y_pos,objects)
plt.xlabel=('paid average')
plt.ylabel=('free average')
plt.show()
end = time.time()
print(end-start)
```
